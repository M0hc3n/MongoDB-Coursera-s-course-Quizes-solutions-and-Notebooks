```python
from pymongo import MongoClient, DESCENDING
import pprint

client = MongoClient(course_cluster_uri)

db = client.sample_mflix

```

```python
filters = {}
page = 0
movies_per_page = 20
```

```python
# key to the value that we will be sorting depending on it
sort_key = "tomatoes.viewer.numReviews"

# we filter the movies' collection then we sort the result depending on number of reviews
# in order to get the most watched movies
# at first, filters will be blank, hence we will be taking the whole collection into consideration
movies_to_be_displayed = db.movies.find(filters).sort(sort_key, DESCENDING)
```

```python
# in order to implement pagination we will need to 
# limit of on screen movies list to only 20 movies
# and at the same time skip the list of movies that the user had seen
movies_on_screen = movies_to_be_displayed.skip(page * movies_per_page).limit(movies_per_page)
```

```python
list_of_movies_on_screen = list(movies_on_screen)

pprint.pprint(list_of_movies_on_screen)

print(len(list_of_movies_on_screen))

# on screen, the top 20 movie right then
```

```python
# we simulate the action of going to the second page
page = 1
```

```python

movies_to_be_displayed = db.movies.find(filters).sort(sort_key, DESCENDING).skip(movies_per_page * page).limit(movies_per_page)

list_of_movies_on_screen = list(movies_to_be_displayed)

pprint.pprint(list_of_movies_on_screen)
```
