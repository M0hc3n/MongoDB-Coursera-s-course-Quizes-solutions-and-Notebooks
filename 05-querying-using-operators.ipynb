```python
from pymongo import MongoClient, DESCENDING
import pprint
```

```python
# Replace XXXX with your connection URI from the Atlas UI
client = MongoClient(XXX)

mflix_movies = client.sample_mflix.movies
```

```python
# $gte means greater than 
# we are using it here to fing all movies greater than 2000

filters = {
    'year': { '$gte': 2000}
}

for movie in mflix_movies.find(filters):
    pprint.pprint(movie['title'])
```

```python
# grabing movies between 2000 and 2009

filters = {
    'year': {'$gte': 2000 , '$lt': 2010} 
}

for movie in mflix_movies.find(filters).sort("tomatoes.viewer.numReviews", DESCENDING):
    pprint.pprint(movie['title'])

# grabing movies between 2000 and 2009 sorted according to the most viewed ones
# for movie in mflix_movies.find(filters).sort("tomatoes.viewer.numReviews", DESCENDING):
#     pprint.pprint(movie['title'])
```

```python
# same example but with using logical ' and ' operator 
filters = {
    '$and': [{ 'year':  {'$gte': 2000} }, { 'year':  {'$lt': 2010}}]
}

for movie in mflix_movies.find(filters):
    pprint.pprint(movie['title'])
```

```python
# grabing movies that were released in 2010 and 2015
filters = {
    'year': { '$in': [2010, 2015]}
}

for movie in mflix_movies.find(filters):
    pprint.pprint(movie['title'])
```

```python
# grabing non-adult movies that were released in 2010 and 2015
# '$ne' stands for not equal

filters = {
    'year': { '$in': [2010, 2015]} , 'genre': { '$ne': 'adult'}
}

for movie in mflix_movies.find(filters):
    pprint.pprint(movie['title'])
```
