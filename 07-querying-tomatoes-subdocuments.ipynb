```python
import pprint
import pymongo
```

```python
# Replace XXXX with your connection URI from the Atlas UI
client = pymongo.MongoClient(XXX)
movies = client.sample_mflix.movies
```

```python
# finding titanic movie
titanic = list(movies.find({'title': "Titanic"}))

pprint.pprint(titanic)
```

```python
# exploring the tomatoes' field
pprint.pprint(titanic[0]['tomatoes'])
```

```python
# we can see how the subdocument is also nested
pprint.pprint(titanic[0]['tomatoes']["viewer"])
```

```python
rating = titanic[0]['tomatoes']["viewer"]['rating']

print(rating)
```

```python
# grabbing all movies that have the same rating as titanic

projection = {
    'title': 1,
    '_id': 0
}

same_as_titanic = list(movies.find({"tomatoes.viewer.rating": rating}, projection))

pprint.pprint(same_as_titanic)
```

```python
# sorting these movies according to last updated date

# updated the projection to get the "tomatoes.lastUpdated" value
projection = {
    'title': 1,
    '_id': 0,
    'tomatoes.lastUpdated': 1,
}

same_as_titanic_sorted = list(movies.find({"tomatoes.viewer.rating": rating}, projection).sort('tomatoes.lastUpdated', pymongo.ASCENDING))

# printing them
for movie in same_as_titanic_sorted:
    pprint.pprint("Title: {0}; lastUpdated: {1}".format(movie['title'], movie['tomatoes']['lastUpdated']))
```
