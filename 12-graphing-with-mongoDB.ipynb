```python
import pymongo
import pprint
```

```python
# Replace XXXX with your connection URI from the Atlas UI
client = pymongo.MongoClient(XXX)
movies = client.sample_mflix.movies
```

```python
query = {
    "runtime": {"$exists": True},
    "metacritic": {"$exists": True}
}

# displaying only runtime and metacritic
projection = {
    '_id':0,
    'runtime':1,
    'metacritic': 1,
}
```

```python
# casting the result into a python list
rm = list(movies.find(query, projection))
```

```python
# displaying a sample of the result
pprint.pprint(rm[0])
```

```python
# extracting a list of runtimes
runtimes = [movie['runtime'] for movie in rm]
```

```python
pprint.pprint(runtimes)
```

```python
# doing the same thing for metacritic ratings
metacritic_ratings = [movie["metacritic"] for movie in rm]
```

```python
pprint.pprint(metacritic_ratings)
```

```python
plt.clf()

fig , ax = plt.subplots() 

# the alpha parameter is used to handle the opacity
ax.scatter(runtimes, metacritic_ratings, alpha=0.5)

# adding the labels for our resulting graph
plt.title("Metacritic Movie Ratings vs. Movie Runtime")
plt.xlabel('Movie Runtime (minutes)')
plt.ylabel('Movie Rating (metacritic)')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
# we will using 3D projecting by adding the year axis
query = {
    "runtime": {"$exists": True},
    "metacritic": {"$exists": True},
    "year": {"$exists": True}
}

projection = {
    '_id':0,
    'runtime':1,
    'metacritic':1,
    'year':1,
}
```

```python
rmy = list(movies.find(query, projection))
```

```python
runtimes = [ float(movie['runtime']) for movie in rmy]
metacritic_ratings = [ float(movie['metacritic']) for movie in rmy]
years = [float(movie['year'])  for movie in rmy]
```

```python
plt.clf()

fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.scatter(runtimes, metacritic_ratings, years)

plt.title("Runtimes, Metacritics and Years")
ax.set_xlabel("Runtimes")
ax.set_ylabel("Metacritic")
ax.set_zlabel("Years")

plt.show()
```

```python
# exploring the pings collection
pprint.pprint(list(pings.find().limit(10)))
```

```python
cursor = pings.aggregate([
    {
        # selecting 5000 random sample from the pings collection
        "$sample": {"size": 5000},
    },
    {
        "$addFields":{
            "dayOfWeek": {"$dayOfWeek": "$ts"}, # returns the day of week corresponding to the datetime object
            "hourOfDay": {"$hour": "$ts"} # returns the hour corresponding to the datetime object
        }
        # adding these fields that we will need later on
    },
    {
        "$group": {"_id": "$dayOfWeek", "pings": {"$push": "$$ROOT"} }
        # we will group the results in document that contains an id which is the day of the week (ex ts)
        # the new field "pings" is an array of documents with grouped based on the id
        # which is (in our case) the date 
        # {"$push": "$$ROOT"} will push any item with similar id to the pings array
    },
    {
        # sorting ascendingly according to the id (i.e: according to the date)
        "$sort":{ "_id": 1 }
    }
])

# the following aggregation returns an array of an array of groups (classified according to the dates)
```

```python
pings_by_day = [group['pings'] for group in cursor]
# this will return the array of groups (of dictionnaries)

pprint.pprint(pings_by_day[0])
```

```python
# getting an array of hourOfDay from the result of the aggregation
pings_by_hour_by_day = [ [ping['hourOfDay'] for ping in pings] for pings in pings_by_day ]

# as if we have gathered all the 'hourOfDay' values into a one array
pprint.pprint(pings_by_hour_by_day[0])
```

```python
# drawing the box plot
plt.clf()

fig, ax = plt.subplots()

ax.boxplot(pings_by_hour_by_day, notch= True)

ax.set_title('When People Watch Movies')
ax.yaxis.grid(True)
ax.set_xticklabels(['Sun', 'Mon', 'Tues', 'Wed', 'Thur', 'Fri', 'Sat'])
ax.set_xlabel('Day of Week')
ax.set_ylabel('Hour of Day')

plt.show()
```
